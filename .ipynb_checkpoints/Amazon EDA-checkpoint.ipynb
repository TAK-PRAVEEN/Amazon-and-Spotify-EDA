```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('amazon.csv')
df.head().T
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
df.columns
```

```python
df['discounted_price'] = df['discounted_price'].str.replace("₹","").str.replace(",","")
df['discounted_price'] = df['discounted_price'].astype(float)
df['discounted_price']
```

```python
df['actual_price'] = df['actual_price'].str.replace("₹","").str.replace(",","")
df['actual_price'] = df['actual_price'].astype(float)
df['actual_price']
```

```python
df['discount_percentage'] = df['discount_percentage'].str.replace("%","")
df['discount_percentage'] = df['discount_percentage'].astype(int)
df['discount_percentage']
```

```python
df.drop(axis=1, index=1279, inplace=True)
```

```python
df['rating'] = df['rating'].astype(float)
df['rating']
```

```python
df['rating_count'] = df['rating_count'].str.replace(",","")
df['rating_count'] = df['rating_count'].astype(float)
df['rating_count']
```

```python
df.info()
```

1. What is the average rating for each product category?

```python
df['category'].value_counts()
```

```python
for i in df['category'].unique():
    average_rating = np.average(df[df['category'] == i]['rating'])
    print(f"{i}\t:\t{average_rating}")
```

2. What are the top rating_count products by category?

```python
df.groupby(['category'])['rating_count'].max().sort_values(ascending=False).reset_index().head(10)
```

3.  What is the distribution of discounted prices vs. actual prices?

```python
fig = px.scatter(data_frame=df, x='discounted_price', y='actual_price', color='rating')
fig.show()
```

```python
sns.scatterplot(data=df, x='discounted_price', y='actual_price', hue='actual_price')
plt.legend(labels=[])
plt.show()
```

4.  How does the average discount percentage vary across categories?

```python
for i in df['category'].unique():
    average_discount = np.average(df[df['category'] == i]['discount_percentage'])
    print(f"{i}\t:\t{average_discount}")
```

8.  What is the correlation between discounted_price and rating?

```python
corr = df[['discounted_price', 'rating']].corr()
corr
```

```python
cmap = sns.color_palette('Spectral')
sns.heatmap(corr, annot=True, cmap=cmap, linecolor='w', linewidths=10)
```
